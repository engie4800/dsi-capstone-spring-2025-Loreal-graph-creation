```python
from tqdm import tqdm
import pandas as pd
from datasets import Dataset
import matplotlib.pyplot as plt
import numpy as np
from datasets import load_dataset
```

```python
ds = load_dataset("Studeni/AMAZON-Products-2023")
```

```python
train_ds = ds["train"]
train_ds
```

```python
df = pd.DataFrame(train_ds)
beauty_df = df[((df['main_category'] == 'Health & Personal Care') | (df['main_category'] == 'All Beauty'))]
beauty_df.head()
```

```python
beauty_df.reset_index(drop=True, inplace=True)
beauty_df.drop(columns="__index_level_0__", inplace=True)
```

```python
beauty_df
```

```python
beauty_df.to_csv("beauty_amazon_dataset.csv", index=False)
```

```python
beauty_df.info()
```

```python
batch_size = 100
file_count = 1
file_name = f'unstructured{file_count}.md'

with open(file_name, 'w') as file:
    for i, row in beauty_df.iterrows():
        
        def clean_text(text):
            return str(text).replace('\u2028', '').replace('\u2029', '').replace('\r', '').replace('\n', ' ').strip()

        new_line = (
            "Title: " + clean_text(row['title']) + "\n" +
            "Description: " + clean_text(row['description']) + "\n" +
            "Date First Available: " + clean_text(row['date_first_available']) + "\n" +
            "Main Category: " + clean_text(row['main_category']) + "\n" +
            "Store: " + clean_text(row['store']) + "\n" +
            "Average Rating: " + clean_text(row['average_rating']) + "\n" +
            "Price: " + clean_text(row['price']) + "\n" +
            "Details: " + clean_text(row['details']) + "\n\n"
        )
        
        file.write(new_line)
        
        if (i + 1) % batch_size == 0:
            file_count += 1
            file_name = f'unstructured{file_count}.md'
            file.close()
            file = open(file_name, 'w')
```
