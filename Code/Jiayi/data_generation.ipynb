```python
from datasets import load_dataset
import pandas as pd
```

```python
ds = load_dataset("milistu/AMAZON-Products-2023")
```

```python
df = ds["train"].to_pandas() # Access the desired split (e.g., "train", "test")
```

```python
df.head()
```

```python
df['main_category'].unique()
```

```python
category_name = "AMAZON FASHION"
selected_df = df[df["main_category"]==category_name]
selected_df = selected_df.drop(columns=["parent_asin", "filename", "embeddings", "image", "__index_level_0__"])
print(f"Number of rows before dropping NA: {len(selected_df)}")
print(f"NA values by columns:")
print(selected_df.isna().sum())
selected_df = selected_df.dropna()
print(f"Number of rows after dropping NA: {len(selected_df)}")
```

```python
store_name = sorted(selected_df["store"].unique())
print(len(store_name))
for store in store_name:
    print(store)
```

```python
samples_df = selected_df[selected_df["store"].isin(["Nike"])]
samples_df = samples_df.sort_values(by="rating_number", ascending=False)
nike_df = samples_df.iloc[0:3]
```

```python
samples_df = selected_df[selected_df["store"].isin(["adidas"])]
samples_df = samples_df.sort_values(by="rating_number", ascending=False)
adidas_df = samples_df.iloc[0:2]
```

```python
save_df = pd.concat([nike_df, adidas_df], ignore_index=True)
save_df
```

```python
DATA_SAVE_PATH = "./data/five_samples.csv"
save_df.to_csv(DATA_SAVE_PATH)
```
